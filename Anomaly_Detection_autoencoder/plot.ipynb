```python
import librosa
import librosa.display
import matplotlib.pyplot as plt
import os

def plot_machine_audio_comparison():

    valve_path = "data/valve/id_00/abnormal/00000002.wav"
    pump_path  = "data/pump/id_00/abnormal/00000005.wav"
    fan_path   = "data/fan/id_00/abnormal/00000001.wav"

    for path in [valve_path, pump_path, fan_path]:
        if not os.path.exists(path):
            print(f"Error: Could not find file at {path}. Please update the filename!")
            return

    y_pump, sr_pump   = librosa.load(pump_path, sr=None)
    y_fan, sr_fan     = librosa.load(fan_path, sr=None)
    y_valve, sr_valve = librosa.load(valve_path, sr=None)


    fig, axes = plt.subplots(nrows=3, ncols=1, figsize=(14, 10), sharex=True, sharey=True)

    librosa.display.waveshow(y_pump, sr=sr_pump, ax=axes[0], color='blue', alpha=0.7)
    axes[0].set_title('Pump (Continuous Drone - Global Mean Works)', fontsize=14, fontweight='bold')
    axes[0].set_ylabel('Amplitude', fontsize=12)

    librosa.display.waveshow(y_fan, sr=sr_fan, ax=axes[1], color='green', alpha=0.7)
    axes[1].set_title('Fan (Continuous Drone - Global Mean Works)', fontsize=14, fontweight='bold')
    axes[1].set_ylabel('Amplitude', fontsize=12)

    librosa.display.waveshow(y_valve, sr=sr_valve, ax=axes[2], color='red', alpha=0.8)
    axes[2].set_title('Valve (Impulsive Clicks - Global Mean Fails)', fontsize=14, fontweight='bold')
    axes[2].set_ylabel('Amplitude', fontsize=12)
    axes[2].set_xlabel('Time (seconds)', fontsize=12)

    plt.suptitle("Acoustic Physics: Continuous vs. Impulsive Machine Failures", fontsize=18)
    plt.tight_layout()
    plt.show()

if __name__ == "__main__":
    plot_machine_audio_comparison()
```
